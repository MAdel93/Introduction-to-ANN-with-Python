```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.load("/content/drive/MyDrive/ANN-HW3/bunny.npy")
```

```python
plt.scatter(data[:, 0], data[:, 1], data[:, 2],label='Data')
plt.title('Dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
neurons = np.random.normal(np.mean(data), np.std(data), size=(1000, data.shape[1]))
neurons
```

```python
plt.scatter(data[:, 0], data[:, 1], data[:, 2],label='Data')
plt.scatter(neurons[:, 0], neurons[:, 1],  neurons[:, 2], label='Neurons')
plt.title('Created Dataset + Neurons Initial Position')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
nEpochs = 50
LR = 0.5 
dLR = -0.01 
Theta = 1/100 
```

```python
nK=1000
```

```python
learning=0.5
for i in range(100):
    for datas in data:
        D = np.linalg.norm(neurons - datas, axis=1)
        J = np.argmin(D)
        neurons[J] += LR * (datas - neurons[J])
        if J == 0:
            neurons[J+1] += 0.01 * learning * (datas - neurons[J+1])
        elif J == nK-1:
            neurons[J-1] += 0.01 * learning * (datas - neurons[J-1])
        else:
            neurons[J+1] += 0.01 * learning * (datas - neurons[J+1])
            neurons[J-1] += 0.01 * learning * (datas - neurons[J-1])
    #LR += dLR
    print(f'Epoch {i+1} Ended.')
```

```python
plt.scatter(data[:, 0], data[:, 1], data[:, 2],label='Data')
plt.scatter(neurons[:, 0], neurons[:, 1],  neurons[:, 2], label='Neurons')
plt.title('Created Dataset + Final Neurons Position After 100 Epoch')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
